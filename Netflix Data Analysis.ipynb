Data Sorting and Cleaning

```python
sorted_nf = nf.sort_values('release_year', ascending = True, kind= 'heapsort', na_position= 'last')
sorted_nf.head()
```

```python
new_nf = sorted_nf.dropna()
```

```python
new_nf.head(5)
```

```python
new_nf.info()
new_nf.shape
```

```python
print(new_nf.duplicated())
```

```python
new_nf.drop_duplicates()
```
